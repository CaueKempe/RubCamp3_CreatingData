```python


# Gerar anos de nascimento usando os pesos normalizados
years_of_birth = np.random.choice(years, size=num_users, p=weights)

# Função para decidir datas de nascimento com base nos anos
def decide_date_of_birth(year: int):
    start_date = date(year, 1, 1)
    end_date = date(year, 12, 31)
    return fake.date_between(start_date=start_date, end_date=end_date)

decide_date_of_birth_vectorized = np.vectorize(decide_date_of_birth)
dates_of_birth = decide_date_of_birth_vectorized(years_of_birth)
display("Datas de nascimento:", dates_of_birth)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Função para gerar dados de acessos e seguidores
def generate_instagram_data(total_views, total_followers, multiplier_september_views=2, multiplier_september_followers=1.5):
    """
    Gera dados fictícios para acessos ao perfil do Instagram e crescimento de seguidores ao longo de 2023.
    
    Parameters:
    - total_views (int): Número total de acessos ao perfil para o ano.
    - total_followers (int): Número total de seguidores ao final do ano.
    - multiplier_september_views (float): Multiplicador para aumentar os acessos em setembro.
    - multiplier_september_followers (float): Fator para aumentar o crescimento de seguidores em setembro.
    
    Returns:
    - pd.DataFrame: DataFrame com os meses, acessos ao perfil e número de seguidores.
    """
    months = pd.date_range(start='2023-01-01', end='2023-12-01', freq='MS').strftime('%B').tolist()
    
    # Distribuir acessos com crescimento linear
    base_views = total_views / 13  # Média de views mensais com setembro em conta
    views = [base_views * (1 + i / 12) for i in range(12)]  # Crescimento linear
    
    # Ajustar setembro
    views[8] = views[8] * multiplier_september_views  # Setembro com o dobro dos acessos
    
    # Normalizar para total_views
    sum_views = sum(views)
    views = [v * total_views / sum_views for v in views]
    
    # Crescimento de seguidores ao longo do tempo
    growth_rate = (total_followers / 1000) ** (1 / 11)  # Taxa de crescimento mensal
    followers = [1000]  # Seguidores base no início do ano
    for i in range(1, 12):
        if i == 8:  # Setembro
            followers.append(followers[-1] * growth_rate * multiplier_september_followers)
        else:
            followers.append(followers[-1] * growth_rate)
    followers = np.round(followers).astype(int)
    
    # Criando o DataFrame
    df_instagram_data = pd.DataFrame({
        'Month': months,
        'ProfileViews': np.round(views).astype(int),
        'Followers': followers
    })
    return df_instagram_data

# Parâmetros para geração dos dados
total_views = 1464895  # Visualizações totais atuais
total_followers = 486622  # Seguidores totais atuais

# Gerando o DataFrame com total_views e total_followers ajustados
df_instagram_data = generate_instagram_data(total_views, total_followers)

# Visualizando o DataFrame
print(df_instagram_data)

# Plotando a distribuição dos acessos ao perfil ao longo do ano
plt.figure(figsize=(12, 8))

# Plot para Profile Views
plt.subplot(2, 1, 1)
plt.bar(df_instagram_data['Month'], df_instagram_data['ProfileViews'], color='skyblue')
plt.title('Acessos ao Perfil do Instagram por Mês em 2023')
plt.xlabel('Mês')
plt.ylabel('Número de Acessos')
plt.xticks(rotation=45, ha='right')

# Plot para Followers
plt.subplot(2, 1, 2)
plt.plot(df_instagram_data['Month'], df_instagram_data['Followers'], marker='o', color='orange')
plt.title('Crescimento de Seguidores no Instagram por Mês em 2023')
plt.xlabel('Mês')
plt.ylabel('Número de Seguidores')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
